```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Handle outliers using the IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for all numerical columns
for column in ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']:
    data = remove_outliers(data, column)

# 2. Normalize the data (except the categorical columns)
scaler = StandardScaler()
numerical_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# 3. Convert categorical columns if needed
# No additional categorical transformation needed as Channel and Region are already numerical
# If needed, you can convert them into 'category' type
data['Channel'] = data['Channel'].astype('category')
data['Region'] = data['Region'].astype('category')

# 4. Double-check for any missing data
missing_data = data.isnull().sum()
print("Missing data after cleaning:\n", missing_data)

# Display the transformed data
print("Cleaned and Transformed Data:\n", data.head())

```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
from sklearn.cluster import KMeans

# Define a function to visualize K-Means with different number of clusters
def kmeans_plot(n_clusters, x, y):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    data['labels_kmeans'] = kmeans.fit_predict(data[[x, y]])
    plot(data[x], data[y], data['labels_kmeans'], f'K-Means with {n_clusters} Clusters: {x} vs {y}')

# Visualize K-Means with different number of clusters (using Grocery vs Fresh)
for n in [2, 3, 4, 5]:
    kmeans_plot(n, 'Grocery', 'Fresh')
```

**Your comment here**

From the scatter plots generated for different cluster numbers (n_clusters = 2, 3, 4, 5), we can visually analyze how well the K-Means algorithm grouped the data.

2 Clusters: This grouping appears overly simplistic, merging too many distinct data points into only two categories.
3 Clusters: A more balanced distribution begins to appear, with clearer separation between the groups.
4 Clusters: This cluster number shows distinct separation and may capture the structure of the data better.
5 Clusters: More clusters could lead to over-segmentation, with some clusters appearing very small or fragmented.
From this experiment, 3 or 4 clusters seem to work best for this dataset.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
from sklearn.cluster import DBSCAN

# Define a function to visualize DBSCAN with different eps and min_samples parameters
def dbscan_plot(eps, min_samples, x, y):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    data['labels_dbscan'] = dbscan.fit_predict(data[[x, y]])
    plot(data[x], data[y], data['labels_dbscan'], f'DBSCAN with eps={eps}, min_samples={min_samples}: {x} vs {y}')

# Visualize DBSCAN with different eps and min_samples values (using Grocery vs Fresh)
for eps, min_samples in [(0.3, 5), (0.5, 10), (0.7, 5), (1.0, 15)]:
    dbscan_plot(eps, min_samples, 'Grocery', 'Fresh')
```

**Your comment here**

The scatter plots above visualize the effects of changing the eps (epsilon) and min_samples parameters in DBSCAN while clustering Grocery vs. Fresh. Here's a summary of the differences:

eps=0.3, min_samples=5: This results in very tight clusters, with many data points being classified as noise.
eps=0.5, min_samples=10: More points are clustered together, with fewer noise points compared to the first setting.
eps=0.7, min_samples=5: With a higher epsilon, the clusters are larger, and even fewer points are considered noise.
eps=1.0, min_samples=15: The highest eps produces the largest clusters, though some denser points may still form distinct clusters.
The choice of eps and min_samples can greatly impact DBSCAN's sensitivity to noise and how it groups the data.
